```python
# Compute Z-scores for numerical columns
z_scores = np.abs(zscore(df.select_dtypes(include=[np.number])))

# Set a threshold for identifying outliers
threshold = 3
outliers = np.where(z_scores > threshold)

print("Outlier positions:", outliers)
```

The outliers variable contains the indices of the data points in the numerical columns of the DataFrame where the absolute Z-scores exceed the threshold (3 in this case). This indicates that these data points are potential outliers because their values are more than three standard deviations away from the mean. The output is a tuple, where the first array represents the row indices and the second array represents the column indices of the detected outliers.

```python
outliers_count = np.sum(outliers, axis = 0)
print("Number of outliers in each column:\n", dict(zip(numerical_cols, outliers_count)))
```

```python
# Remove rows where any Z-score is above the threshold
df_no_outliers = df[(z_scores < threshold).all(axis=1)]

print("Original data shape:", df.shape)
print("Data shape after removing outliers:", df_no_outliers.shape)
```

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns

# Compute Z-scores
z_scores = np.abs(zscore(df[numerical_cols]))

# Set threshold for identifying outliers
threshold = 3

# Cap outliers at threshold
for col in numerical_cols:
    col_mean = df[col].mean()
    col_std = df[col].std()
    lower_limit = col_mean - threshold * col_std
    upper_limit = col_mean + threshold * col_std

    # Cap values
    df[col] = np.where(df[col] < lower_limit, lower_limit,
                       np.where(df[col] > upper_limit, upper_limit, df[col]))

# Verify capping
print("Outliers capped in the dataset")
```

This code identifies and caps outliers in the numerical columns of the DataFrame. For each column, values falling outside the range defined by the mean Â± 3 standard deviations are replaced with the nearest boundary (lower or upper limit). As a result, extreme values are effectively handled, ensuring the dataset is less skewed and more robust to outliers.

```python
# Hyperparameter Tuning: Decision Tree Regressor
dt_param_grid = {'max_depth': [5, 10, 20], 'min_samples_split': [2, 5, 10]}
dt_grid = GridSearchCV(DecisionTreeRegressor(random_state=42), dt_param_grid, cv=5, scoring='r2')
dt_grid.fit(X_train, y_train)
best_dt = dt_grid.best_estimator_

# Hyperparameter Tuning: Random Forest Regressor
rf_param_grid = {'n_estimators': [50, 100, 200], 'max_depth': [5, 10, 20], 'min_samples_split': [2, 5, 10]}
rf_grid = GridSearchCV(RandomForestRegressor(random_state=42), rf_param_grid, cv=5, scoring='r2')
rf_grid.fit(X_train, y_train)
best_rf = rf_grid.best_estimator_

# Hyperparameter Tuning: Gradient Boosting Regressor
gb_param_grid = {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2], 'max_depth': [3, 5, 10]}
gb_grid = GridSearchCV(GradientBoostingRegressor(random_state=42), gb_param_grid, cv=5, scoring='r2')
gb_grid.fit(X_train_scaled, y_train)
best_gb = gb_grid.best_estimator_

# Model Definitions
models = {
    'Linear Regression': Pipeline([('scaler', StandardScaler()), ('model', LinearRegression())]),
    'Support Vector Regressor': Pipeline([('scaler', StandardScaler()), ('model', SVR())]),
    'Best Decision Tree': best_dt,
    'Best Random Forest': best_rf,
    'Best Gradient Boosting': best_gb
}

# Cross-validation and model evaluation
for model_name, model in models.items():
    if 'Pipeline' in str(type(model)):  # Scale inside pipeline for linear models and SVR
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
    
    # Cross-validated RÂ² score
    scores = cross_val_score(model, X_train_scaled if 'Pipeline' in str(type(model)) else X_train, y_train, cv=5, scoring='r2')
    mean_cv_r2 = scores.mean()
    
    # Metrics
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"\nMetrics for {model_name}:")
    print(f"Cross-validated RÂ²: {mean_cv_r2}")
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")
    print(f"RÂ² Score: {r2}")
```
